# Model Evaluation and Refinement


Import the necessary modules:


```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
```

We will split the data into training and testing sets:


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```


Creating and fitting a Ridge regression object using the training data, setting the regularization parameter to 0.1, and calculating the R^2 using the test data.


```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
y_pred = ridge.predict(x_test)
r2 = r2_score(y_test, y_pred)
print("R^2 score:", r2)
```


Performing a second order polynomial transform on both test and traing data. Creat and fit a Ridge regression object using the training data, set the regularisation parameter to 0.1, and calculate the R^2 utilising the test data provided. Take a screenshot of your code and the R^2.


```python
poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
reg = Ridge(alpha=0.1)
reg.fit(x_train_poly, y_train)
y_pred = reg.predict(x_test_poly)
r2 = r2_score(y_test, y_pred)
print("R^2:", r2)
```

# INSIGHTS


Square footage of living area has the highest positive correlation (0.70) with price, indicating larger homes sell for more. This aligns with findings that home size is a key determinant of price (Hansz and Diaz III, 2001).

Number of bathrooms (0.53) and bedrooms (0.31) also have moderately high positive correlations, suggesting more baths and bedrooms add value. Research shows additional bedrooms and bathrooms increase resale value (Gallinelli, 2019).

View (0.40) and waterfront (0.27) have positive correlations, showing that good views and waterfront locations command higher prices. Studies confirm water views (Benson et al., 1998) and mountain vistas (Bourassa et al., 2004) positively impact prices.
Overall, size, amenities and location attributes like view are key drivers of home values based on this correlation analysis. Structural factors like age appear less important. This aligns with literature on the predominance of physical and locational attributes in pricing (Sirmans et al., 2005).

The R^2 between the 'long' feature (longitude) and 'price' is low at 0.0005, indicating almost no linear relationship. This aligns with findings that macro location factors like longitude have little direct predictive power on home prices (Watkins, 2001).However, the R^2 between 'sqft_living' (square feet of living space) and 'price' is moderately high at 0.49.
This indicates a fairly positive linear relationship, where larger living spaces correlate with higher prices. This matches extensive research showing living area is a major driver of home values (Sirmans et al., 2005; Hansz and Diaz III, 2001).

Fitting a linear model using multiple features produced an R^2 of 0.66, indicating a moderately strong relationship between the input features and the target home prices. This is substantially higher than using only sqft_living alone (R^2 of 0.49 in the previous analysis). 

Overall, this multiple regression model effectively captures the effect of key pricing factors, demonstrating the importance of using multiple predictors rather than single variables for modeling housing values.
Splitting the data into 85% training and 15% testing sets is an appropriate ratio for fitting and evaluating machine learning models (Arlot and Celisse, 2010).The baseline Ridge regression model achieved an R^2 of 0.65 on the test data. This indicates moderately good but not excellent out-of-sample predictive performance.

Transforming the features by applying 2nd order polynomial expansion improved the R^2 to 0.70. This aligns with literature suggesting nonlinear transformations can enhance model flexibility and accuracy (Lipovetsky and Conklin, 2001).
The model improvement from 0.65 to 0.70 R^2 seems minor but can be highly meaningful.
Further refinements like regularizing to reduce overfitting, adding interaction variables and incorporating macroeconomic data could likely improve accuracy even more (Park and Bao, 2015).
