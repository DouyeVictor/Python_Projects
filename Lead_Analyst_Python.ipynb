```python
#importing CSV files
Client_Table = pd.read_csv('Client_Table.csv')
Date_Table = pd.read_csv('Date_Table.csv')
Long_Term_Support_Table = pd.read_csv('Long_Term_Support_Table.csv')
New_Admissions_Table = pd.read_csv('New_Admissions_Table.csv')
Request_Table = pd.read_csv('Request_Table.csv')
```

```python
#Data Inspection and Cleaning

# Handling missing values
Client_Table = Client_Table.dropna()
Date_Table = Date_Table.dropna()
Long_Term_Support_Table = Long_Term_Support_Table.dropna()
New_Admissions_Table = New_Admissions_Table.dropna()
Request_Table = Request_Table.dropna()

# Dropping Duplicates
Client_Table = Client_Table.drop_duplicates()
Date_Table = Date_Table.drop_duplicates()
Long_Term_Support_Table = Long_Term_Support_Table.drop_duplicates()
New_Admissions_Table = New_Admissions_Table.drop_duplicates()
Request_Table = Request_Table.drop_duplicates()
```

```python
# Viewing cleaned data to confirm consistency
Client_Table.head(6)
```

```python
# I noticed 5 extra Coulumns named 'unnamed'
# Deleting irrelevant columns
Client_Table = Client_Table.drop(Client_Table.columns[2:7], axis=1)  # Delete 3rd to 7th columns
Client_Table.head(6)
```

```python
# Deleting rows with '0'
New_Admissions_Table.drop(New_Admissions_Table[New_Admissions_Table['EndDate'] == 0].index, inplace=True)
```

```python
#Saving Cleaned Tables as CSV files for joining in SQL
Client_Table.to_csv('Clean_Client_Table.csv', index=False, header=True)
Date_Table.to_csv('Clean_Date_Table.csv', index=False, header=True)
New_Admissions_Table.to_csv('Clean_New_Admissions_Table.csv', index=False, header=True)
Long_Term_Support_Table.to_csv('Clean_Long_Term_Support_Table.csv', index=False, header=True)
Request_Table.to_csv('Clean_Request_Table.csv', index=False, header=True)
```
